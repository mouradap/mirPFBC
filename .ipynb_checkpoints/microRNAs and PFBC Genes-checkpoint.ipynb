```python
import pandas as pd
```

This projects aims to analyze the microRNAs involved with Primary Familial Brain Calcification causing genes.<br>
For this, we are going to use the miRWalk database (http://mirwalk.umm.uni-heidelberg.de/about/) hits on the six genes, SLC20A2, PDGFB, PDGFRB, XPR1, MYORG and JAM2.<br>

```python
df = pd.read_csv('miRWalk_miRNA_Targets.csv')
df.head()
```

miRWalk database includes probable microRNA interactions with each gene transcript. <br>
It contains information about several miR-DNA binding sites, such as:<br>

mirnaid: miR ID<br>
refseqid: Transcript ID<br>
genesymbol: Gene Symbol<br>
start: Initial binding site<br>
end: Final binding site<br>
bindinggp: ?<br>
energy: Folding Energy<br>
seed: Seed match<br>
accessibility: Accessibility<br>
au: AU content. Au shows the AU-rich elements (AREs). The local AU content reflects the transcript AU content 30nt upstream and downstream of predicted site.<br>
phylopstem:        Stem Conservation<br>
phylopflank:        Flanking conservation<br>
me:        m/e motif. This feature is about the paring probabilities at different positions of miRNA.<br>
number_of_pairings:        Total number of paired positions<br>
binding_region_length:        The length of target mRNA region<br>
longest_consecutive_pairings:        The length of the largest consecutive pairs<br>
position:        Position of the largest consecutive pairs<br>
validated:<br>
TargetScan:<br>
miRDB:<br>

Also,

        Conservation Difference
        The length of the largest consecutive pairs allowing 2 mismatches
        The position of the largest consecutive pairs allowing 2 mismatches
        The number of paired positions at the miRNA 3’ end
        The total number of paired positions in the seed region and the miRNA 3’ end
        The difference between the number of paired positions in the seed region and that in the miRNA 3’ end
        Exon preference

Having understood how the data is presented in this database, we are now going to download data for all the other six PFBC genes.

```python
myorg = pd.read_csv('myorg.csv')
myorg.drop('validated',axis=1,inplace=True)
pdgfb = pd.read_csv('pdgfb.csv')
pdgfb.drop('validated',axis=1,inplace=True)
pdgfrb = pd.read_csv('pdgfrb.csv')
pdgfrb.drop('validated',axis=1,inplace=True)
xpr1 = pd.read_csv('xpr1.csv')
xpr1.drop('validated',axis=1,inplace=True)
jam2 = pd.read_csv('jam2.csv')
jam2.drop('validated',axis=1,inplace=True)
```

```python
combined = df.merge(myorg, how = 'inner', on = ['mirnaid'])
combined = combined.merge(xpr1, how = 'inner', on = ['mirnaid'])
combined = combined.merge(pdgfb, how = 'inner', on = ['mirnaid'])
combined = combined.merge(pdgfrb, how = 'inner', on = ['mirnaid'])
combined = combined.merge(jam2, how = 'inner', on = ['mirnaid'])
```

```python
combined.dropna(inplace=True)
```

```python
combined.head()
```

```python
mirList = combined['mirnaid'].unique()
```
